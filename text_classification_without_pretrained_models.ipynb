```python
import pandas as pd
df=pd.read_excel('input_data.xls')
df.head(10)
```

```python
df.info()
```

```python
dt=df[5:230]
```

```python
dt.head()
```

```python
dt.columns=['Date','Narration','Value Dt','Withdrawal Amt','Deposit Amt']
```

```python
import numpy as np
dp=dt.replace(to_replace = np.nan, value =0)
dp['Deposit Amt'].values[dp['Deposit Amt'] > 0] = 1
```

```python
dp.head(10)
```

**Credits** **Bucket**

```python
creditsdf= dp[dp['Deposit Amt'] == 1]
credits=creditsdf['Narration'].values
credits_num=len(credits)
print(credits_num)
credits
```

**Debits** **Bucket**

```python
debitsdf= dp[dp['Deposit Amt'] == 0]
debits=debitsdf['Narration'].values
debits_num=len(debits)
print(debits_num)
debits
```

**Broker** **Bucket**

```python
broker=[]
for item in debits:
     if 'ZERODHA' in item:
       broker.append(item)
print(len(broker))       
broker 
```

**Onlinefoodpayments Bucket**

```python
online_food=['ZOMATO','SWIGGY']
onlinefoodpayments=[]
for item in debits:
   #items=item.split("-")
   for ele in online_food:
     if ele in item:
       onlinefoodpayments.append(item)
print(len(onlinefoodpayments))       
onlinefoodpayments 
```

**Onlineshoppayments Bucket**

```python
online_shoping=['AMAZON','FLIPKART']
onlineshoppayments=[]
for item in debits:
   #items=item.split("-")
   for ele in online_shoping:
     if ele in item:
       onlineshoppayments.append(item)
print(len(onlineshoppayments))       
onlineshoppayments 
```

**Funds Transfer Bucket**

```python
funds_transfer=[]
for item in debits:
     if 'FUNDS' in item:
       funds_transfer.append(item)
print(len(funds_transfer))       
funds_transfer 
```

**Cash Withdrawal Bucket**

```python
withdraw=['ATW','NWD']
cash_withdrawal=[]
for item in debits:
   #items=item.split("-")
   for ele in withdraw:
     if ele in item:
       cash_withdrawal.append(item)
print(len(cash_withdrawal))       
cash_withdrawal 
```
